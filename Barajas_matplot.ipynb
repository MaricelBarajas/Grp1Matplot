```python
import matplotlib.pyplot as plt

departments = ["A", "B", "C", "D"]
sales = [90, 82, 63, 84]
sales_performance = [90, 82, 63, 84]  # Corrected variable assignment

# Plotting
plt.figure(figsize=(12, 5))  # Corrected assignment operator

# Subplot 1
plt.subplot(1, 2, 1)
plt.bar(departments, sales, color="darkblue")  # Corrected color assignment
plt.title('Sales Performance by Department')
plt.xlabel('Departments')
plt.ylabel('Sales')
plt.grid(axis="y")  # Corrected axis assignment

# Subplot 2
plt.subplot(1, 2, 2)
plt.pie(sales_performance, labels=departments, autopct="%1.1f%%", startangle=148, colors=['blue', 'yellow', 'green', 'red'])  # Corrected autopct and startangle arguments
plt.title('Sales Performance Percentage by Department')
plt.tight_layout()

plt.show()
```
